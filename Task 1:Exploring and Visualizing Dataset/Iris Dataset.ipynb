# **EXPLORING AND VISUALIZING IRIS DATASET VISUALIZATION**
---
The goal of this project is to explore and analyze Iris dataset.It consists of 150 observations of iris flowers, categorized into three different species: Iris-setosa, Iris-versicolor, and Iris-virginica. Each observation includes four numerical features: sepal length, sepal width, petal length, and petal widthâ€”all measured in centimeters.

In this notebook, we explore the dataset using Python, using libraries like Pandas, Matplotlib, and Seaborn to understand the relationships between different flower attributes and species. Through various visualizations such as scatter plots, histograms, and box plots, we aim to:

- Understand feature distributions and relationships.
- Detect outliers and variability in measurements.
- Identify patterns that help distinguish between species.

---
## Dataset Overview
**Source:** https://www.kaggle.com/datasets/saurabh00007/iriscsv/code

**Description:** This dataset contains information about Iris flower.It consist of 6 columns which includes:

- **Id:** Unique identifier for each record
- **SepalLengthCm:** Sepal length in centimeters
- **SepalWidthCm:** Sepal width in centimeters
- **PetalLengthCm:** Petal length in centimeters
- **PetalWidthCm:** Petal width in centimeters
- **Species:** Type of Iris flower (Setosa, Versicolor, Virginica)
---

## **STEP 1: LOADING DATA**

```python
# importing all neccessary libraries i-e pandas to read dataframe,matplotlib and seaborn for visualization
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

# Reading dataset
df=pd.read_csv("iris.csv")

```

## **STEP 2: DISPLAYING DATASET STRUCTURE**

## **Dataset Info**

```python
print(df.info())
```

## **Dataset structure**

```python
print("Shape of Dataset=",df.shape)
print("Columns in dataset=",df.columns.tolist())
print("The first 10 rows of dataset are:\n",df.head(10).to_string(index=False))
```

## **Summary Statistics**

```python
df.describe()
```

## **STEP 3: VISUALIZATION**

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x="SepalLengthCm",y="SepalWidthCm",hue="Species",palette="Set2")
plt.title("SEPAL LENGTH VS SEPAL WIDTH")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x="PetalLengthCm",y="PetalWidthCm",hue="Species",palette="Set1")
plt.title("Petal LENGTH VS Petal WIDTH")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df["PetalLengthCm"],bins=20,kde=True,color="Blue")
plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.tight_layout()
plt.show()
```

## **Box plot to show the spread of outliers across different species**

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Species', y='SepalLengthCm',hue="Species", palette='pastel')
plt.title("Box Plot of Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# **Conclusion**

From our visual exploration, particularly the box plot of Sepal Length, we observe that:

- Iris-virginica generally has the highest sepal length among the three species.
- Iris-setosa has shorter sepal lengths with relatively low variance, making it the most distinct in terms of this feature.
- Iris-versicolor lies between the two, with moderate length and variability.
- There are some outliers in each species, but overall, the feature helps in differentiating between flower types.

This exploration highlights how even simple visual tools can uncover meaningful patterns that might be missed in raw data.
